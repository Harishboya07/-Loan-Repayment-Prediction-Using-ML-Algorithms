# 4.Handling outliers

```python
# Here i have declared a variable name cols1 for columns which have outliers

cols1 = ['int.rate','installment','log.annual.inc','fico','days.with.cr.line','revol.bal']
```

```python
# by using box plot i have generated plots for columns having outliers

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 12))
for i, col in enumerate(cols1[:11]):
  sns.boxplot(df[col], ax=axes[i // 3, i % 3])
  axes[i // 3, i % 3].set_title(col)
plt.tight_layout()
```

```python
# By using IOR(Interqurartile Range) with median imputation i have removed outliers and plots are generated

Q1 = df['int.rate'].quantile(0.25)
Q3 = df['int.rate'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df['int.rate'] = np.where(
    (df['int.rate'] < lower_limit),lower_limit,np.where(
         (df['int.rate'] > upper_limit),upper_limit,
    df['int.rate']
))
sns.boxplot(df['int.rate'])
plt.title("BOX PLOT AFTER MEDIAN IMPUTATION")
plt.show()
```

```python
Q1 = df['installment'].quantile(0.25)
Q3 = df['installment'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df['installment'] = np.where(
    (df['installment'] < lower_limit),lower_limit,np.where(
         (df['installment'] > upper_limit),upper_limit,
    df['installment']
    ))
sns.boxplot(df['installment'])
plt.title("BOX PLOT AFTER MEDIAN IMPUTATION")
plt.show()
```

```python
Q1 = df['log.annual.inc'].quantile(0.25)
Q3 = df['log.annual.inc'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df['log.annual.inc'] = np.where(
    (df['log.annual.inc'] < lower_limit),lower_limit,np.where(
         (df['log.annual.inc'] > upper_limit),upper_limit,
    df['log.annual.inc']
    ))
sns.boxplot(df['log.annual.inc'])
plt.title("BOX PLOT AFTER MEDIAN IMPUTATION")
plt.show()
```

# 7.Declare feature vector and target variable

```python
# x is declared as a variable containing all columns except target column
x=df.drop(['not.fully.paid'],axis=1)

# y is declared as a variable containing only target column
y=df[['not.fully.paid']]
```

# 8.Split dataset to separate training and testing sets

```python
# splittting the data into training and testing sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# 9.SMOTE(synthetic minority over-sampling technique)

```python
# form above histplot and pie chart we know that our data set is imbalance
# so SMOTE technique is applied to balance the data set

from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=0)
x_train_resampled,y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
# below Pie chart we can observe that our datset is balanced now

ax=y_train_resampled.value_counts().plot(kind='pie',autopct='%.2f')
ax
plt.show()
```

```python
# shape of the balanced training data set

x_train_resampled.shape,y_train_resampled.shape
```

```python
# shape of testing data set

x_test.shape,y_test.shape
```

# 10.Applying algorithms

# 10.1.Logistic Regression

```python
# applying logistic regression to dataset

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
lr=LogisticRegression()
```

```python
# Define hyperparameters grid

param_grid = {
    'penalty': ['l1', 'l2'],  # Regularization type
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Inverse regularization strength
    'solver': ['liblinear', 'saga']  # Optimization algorithm
}
```

```python
# Initializing GridSearchCV

grid_search = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')
```

```python
# Fitting x_train_resampled, y_train_resampled to gird search

grid_search.fit(x_train_resampled, y_train_resampled)
```

```python
# After applying different parametrs these are the best parameters

best_params = grid_search.best_params_
best_params
```

```python
# Fit logistic regression with the best parameters

best_lr = LogisticRegression(**best_params)
best_lr.fit(x_train_resampled, y_train_resampled)
```

```python
ypred1=best_lr.predict(x_test)
ypred1
```

```python
# Checking the performance of training

from sklearn.metrics import accuracy_score
lr1=best_lr.score(x_train_resampled,y_train_resampled)*100
lr1
```

```python
# Testing accuracy of logistic Regression model

accuracy1=accuracy_score(y_test,ypred1)*100
accuracy1
```

# confusion Matrix

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
cm_test = confusion_matrix(y_test, ypred1)
matrix = ConfusionMatrixDisplay(confusion_matrix = cm_test, display_labels = [0, 1])
matrix.plot(cmap="Greys")
plt.title("ConfusionMatrix.", weight="bold");
```

# Classification Report

```python
cr1=classification_report(y_test,ypred1)
print(cr1)
```
